```python
import pandas as pd
import numpy as np
```

```python
vehicles_dataset = pd.read_csv('vehicles.csv')
vehicles_dataset
```

```python
vehicles_dataset.columns
```

```python
curr_fleet_np = vehicles_dataset["Current fleet"].to_numpy()
curr_fleet_np
```

```python
sample_means = []
for _ in range(10000):
    #sample size 5
    random_sample_list = np.random.choice(curr_fleet_np, 5)
    sample_means.append(np.mean(random_sample_list))
```

```python
sample_means
```

```python
print(f"Total sample means list size {len(sample_means)}")
print(f"15% data size {(len(sample_means) / 100) * 15}")
```

```python
final_means = sample_means[750:9250]
print(f"Final means list size for 85% {len(final_means)}")
print(f"Final Mean of list {np.mean(final_means)}")
```

```python
print(f"Total sample means list size {len(sample_means)}")
print(f"8% data size {(len(sample_means) / 100) * 8}")
```

```python
final_means = sample_means[400:9600]
print(f"Final means list size for 92% {len(final_means)}")
print(f"Final Mean of list {np.mean(final_means)}")
```

```python
#Standard Deviation
sample_standard_deviation_means = []
for _ in range(10000):
    random_sample_list = np.random.choice(curr_fleet_np, 5)
    sample_standard_deviation_means.append(np.std(random_sample_list))
```

```python
print(f"Total sample standard deviation means list size {len(sample_standard_deviation_means)}")
print(f"15% data size {(len(sample_standard_deviation_means) / 100) * 15}")
```

```python
final_means = sample_standard_deviation_means[750:9250]
print(f"Final means list size for 85% {len(final_means)}")
print(f"Final Mean of list {np.mean(final_means)}")
```

```python
print(f"Total sample standard deviation means list size {len(sample_standard_deviation_means)}")
print(f"15% data size {(len(sample_standard_deviation_means) / 100) * 5}")
```

```python
final_means = sample_standard_deviation_means[250:9750]
print(f"Final means list size for 95% {len(final_means)}")
print(f"Final Mean of list {np.mean(final_means)}")
```

```python
new_fleet_np = vehicles_dataset["New Fleet"].to_numpy()
clean_new_fleet = []
for data in new_fleet_np:
    if not np.isnan(data):
        clean_new_fleet.append(data)
    
clean_new_fleet = np.array(clean_new_fleet)
clean_new_fleet
```

```python
old_fleet_mean = np.mean(curr_fleet_np)
new_fleet_mean = np.mean(clean_new_fleet)

observed_test_result = new_fleet_mean - old_fleet_mean
observed_test_result
```

```python
times_perm_test_more_ext = 0

for _ in range(30000):
    concat_array = np.concatenate((curr_fleet_np, clean_new_fleet))
    perm = np.random.permutation(concat_array)
    pold = perm[:int(len(perm)/2)]
    pnew = perm[int(len(perm)/2):]
    perm_test_result = np.mean(pnew) - np.mean(pold)

    if perm_test_result > observed_test_result:
        times_perm_test_more_ext += 1
```

```python
times_perm_test_more_ext / 30000
```
